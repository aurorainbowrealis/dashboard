```python
"""
We have a pandas dataframe with the following format:

+---+------+------+------+------+------+------+------+------------+--------+---------+
| ID|PAGE_1|PAGE_2|PAGE_3|PAGE_4|PAGE_5|PAGE_6|PAGE_7|DAY_OF_MONTH|HAS_CALL|IS_MOBILE|
+---+------+------+------+------+------+------+------+------------+--------+---------+
|  1|     1|     0|     2|     1|     1|     1|     1|           1|       1|        0|
|  2|     0|     3|     0|     3|     0|     0|     2|           1|       1|        1|
|  5|     7|     0|     0|     0|     4|     2|     0|           1|       1|        0|
|  4|     0|     2|     0|     1|     0|     0|     0|           2|       1|        1|
|  3|     1|     0|     1|     0|     0|     0|     1|           3|       1|        0|
+---+------+------+------+------+------+------+------+------------+--------+---------+

Where id is a unique session id. 
PAGE 1 to N contains the number of times the user visited a page in a session.
DAY_OF_MONTH is the day of the month of the session.
HAS_CALL is a boolean indicating whether the user called the call center within 2h of the session.
IS_MOBILE is a boolean indicating whether the user accessed the site from a mobile device or not.
"""
import pandas as pd
import numpy as np

# making a dataframe
df = pd.DataFrame({
    'ID': [1, 2, 5, 4, 3],
    'PAGE_1': [1, 0, 7, 0, 1],
    'PAGE_2': [1, 3, 0, 2, 0],
    'PAGE_3': [2, 0, 0, 0, 1],
    'PAGE_4': [1, 3, 0, 1, 0],
    'PAGE_5': [1, 0, 4, 0, 0],
    'PAGE_6': [1, 0, 2, 0, 0],
    'PAGE_7': [1, 2, 0, 0, 1],
    'DAY_OF_MONTH': [1, 1, 1, 2, 3],
    'HAS_CALL': [1, 1, 1, 1, 1],
    'IS_MOBILE': [0, 1, 0, 1, 0]
})
df.set_index('ID', inplace=True)

```

```python
# use sdf to calculate the term frequency within each session
# term frequency is the number of times a page was visited during a session
# divided by the total number of visits to all pages in the session
tf = df[df.columns[0:7]].div(df[df.columns[0:7]].sum(axis=1), axis=0)


# use df to calculate the inverse document frequency within each session
# inverse document frequency is the log of the total number of documents
# divided by the number of documents with the word in it
idf = np.log(df.shape[0] / (df[df.columns[0:7]] > 0).sum(axis=0))

# calculate the tf-idf within each session
tf_idf = tf * idf
#normalizing
tf_idf = tf_idf.div(tf_idf.sum(axis=1), axis=0)
```

```python
# create a new dataframe replacing the original values in df with the tf-idf values
dfx = df.copy()
dfx[df.columns[0:7]] = tf_idf

dfx
```
